```python
import pandas as pd

base_path = "/content/drive/MyDrive/Colab Notebooks/Target churn"
persondf = pd.read_parquet(f"{base_path}/clients.parquet")
billdf = pd.read_csv(f"{base_path}/billing.csv", sep=";")
tenuredf = pd.read_json(f"{base_path}/tenure.json", orient="records")
```

```python
persondf["churn_month"] = persondf["churn_date"].dt.month
billdf["billing_date"] = pd.to_datetime(billdf["billing_date"], errors="coerce")
billdf["month"] = billdf["billing_date"].dt.month
tenuredf["month"] = tenuredf["date"].dt.month
df = persondf.merge(billdf, on="customerid").merge(tenuredf, on=["customerid", "month"], how="left")
```

```python
df
```

```python
df["target"] = (df["churn_date"] - df["billing_date"]).dt.days < 100
df["target"] = df["target"].astype(int)
```

```python
df = df.sort_values(["customerid", "month"])
```

```python
df["increased_rate_diff"] = df.groupby("customerid")["monthlycharges"].diff(1)
df["increased_rate_pct"] = df.groupby("customerid")["monthlycharges"].pct_change(1)
```

```python
df["age"] = (df["billing_date"] - df['birth_date']).astype('timedelta64[Y]').astype('int')
df["contract_months"] = (df["billing_date"] - df['contract_date']).astype('timedelta64[M]').astype('int')
df = df.drop(columns=["churn_date", "billing_date", "date", "churn_month"])
```

```python
allowed_digits = ["7", "0", "8", "6", "4", "b"]

# Convertir la lista de dígitos permitidos a una cadena para facilitar la comparación
allowed_digits_str = ''.join(allowed_digits)

# Filtrar el DataFrame para que los últimos dígitos del customerid estén en la lista de dígitos permitidos
filtered_df = df[df['customerid'].str[-1].isin(allowed_digits)]

filtered_df
```

```python
from sklearn.model_selection import train_test_split

# Separación de datos en conjunto de características (X) y variable objetivo (y)
X = filtered_df.drop(columns=['target'])
y = filtered_df['target']

# Identificar los customer_id únicos
unique_customer_ids = filtered_df['customerid'].unique()

# Dividir los customer_id en dos grupos (entrenamiento y holdout)
train_customer_ids, holdout_customer_ids = train_test_split(unique_customer_ids, test_size=0.2, random_state=42)

# Filtrar los datos correspondientes a los customer_id de cada conjunto
train_df = filtered_df[filtered_df['customerid'].isin(train_customer_ids)]
holdout_df = filtered_df[filtered_df['customerid'].isin(holdout_customer_ids)]

# Verificar que los conjuntos sean exclusivos
assert len(set(train_df['customerid']).intersection(set(holdout_df['customerid']))) == 0, "Se encontraron customer_id en ambos conjuntos"

# Verificar la proporción de churn en ambos conjuntos
train_churn_rate = train_df['target'].mean()
holdout_churn_rate = holdout_df['target'].mean()

print("Proporción de churn en conjunto de entrenamiento:", train_churn_rate)
print("Proporción de churn en conjunto de holdout:", holdout_churn_rate)

```
